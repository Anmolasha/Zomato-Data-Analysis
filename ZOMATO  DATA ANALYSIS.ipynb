### Removing "New","_", and "/5" from Rate Column

```python
def newrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value= str(value).split('/')
        value=value[0]
        return float(value)
    
df['rate']= df['rate'].apply(newrate)
df['rate'].head()
        
```

### Filling Null Values in Rate Column with Mean

```python
df['rate'].mode()
```

```python
df['rate']=df['rate'].fillna(df['rate'].mode()[0])
df['rate'].isnull().sum()
```

```python
df.info()
```

### Dropping Null Values

```python
df.rename(columns={'approx_cost(for two people)': 'Cost','listed_in(type)':'Type'},inplace=True)
df
```

### Cleaning Location Column

```python
location=df['location'].value_counts().unique()
location
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#lets simplify the data 

location=df['location'].value_counts(ascending=False)

location_lessthan300= location[location<300]
location_lessthan300

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location']= df['location'].apply(handle_location)
df['location'].value_counts()
    
```

```python
df.isnull().sum()
```

```python
df['location']=df['location'].fillna(df['location'].mode()[0])
df['location']
```

```python
df.isnull().sum()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type

rest=rest_type[rest_type<1000]
rest
```

### Making Rest Types Less than 1000 in frequency as others


```python
def handle_rest_type(value):
    if (value in rest):
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['rest_type']=df['rest_type'].fillna(df['rest_type'].mode()[0])
df['rest_type']
```

```python
df.isnull().sum()
```

### Cleaning Cuisines Column

```python
cuisines=df['cuisines'].value_counts(ascending=False)

cuisines_lessthan100 =cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return 'value'
    
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()

    
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['cuisines']=df['cuisines'].fillna(df['cuisines'].mode()[0])
df['cuisines']
```

### Removing from Cost Column

```python
def handlecomma(value):
    value= str(value)
    if ',' in value:
        value= value.replace(',','')
        return float(value)
    else:
        return float(value)

df['Cost']=df['Cost'].apply(handlecomma)
df['Cost']
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Cost']=df['Cost'].fillna(df['Cost'].mode()[0])
df['Cost']
```
